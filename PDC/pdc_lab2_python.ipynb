PYTHON LIST

```python
import datetime
import numpy as np

print("WITHOUT LOOP M1 = M2 + M3")
t1 = datetime.datetime.now()
print("t1:", t1)

M2 = [[2 for i in range(100)] for j in range(100)]
M3 = [[3 for i in range(100)] for j in range(100)]
M1 = M2 + M3

t2 = datetime.datetime.now()
print("t2:", t2)

print("TIME ELAPSED:", t2-t1)

print("M2:\n", M2)
print("M3:\n", M3)
print("M1:\n", M1)
```

```python
import datetime

print("ROW MAJOR LOOPS R-C")
t1 = datetime.datetime.now()
print("t1:", t1)

M2 = [[2 for i in range(100)] for j in range(100)]
M3 = [[3 for i in range(100)] for j in range(100)]
M1 = [[0 for i in range(100)] for j in range(100)]

for i in range(100):
    for j in range(100):
        M1[i][j] = M2[i][j] + M3[i][j]

t2 = datetime.datetime.now()
print("t2:", t2)

print("TIME ELAPSED:", t2-t1)

print("M2:", M2)
print("M3:" ,M3)
print("M1:",M1)
```

PYTHON - NUMPY

```python
import datetime
import numpy as np

print("NUMPY - WITHOUT LOOP M1 = M2 + M3")
t1 = datetime.datetime.now()
print("t1:", t1)

M2 = np.array([[2 for i in range(100)] for j in range(100)])
M3 = np.array([[3 for i in range(100)] for j in range(100)])
M1 = M2 + M3

t2 = datetime.datetime.now()
print("t2:", t2)

print("TIME ELAPSED:", t2-t1)

print("M2:\n", M2)
print("M3:\n", M3)
print("M1:\n", M1)
```
